```python
from bs4 import BeautifulSoup
from time import sleep
from urllib.request import urlopen
from urllib import parse as urlparse
from dateutil.parser import parse
import pandas as pd

parks_list = ['Iona National Park','Quiçama National Park', 'Cameia National Park', 'Mosi-oa-Tunya National Park', 'Daan Viljoen Game Reserve', 'Karoline National Park']
converter = lambda x: x.replace(' ', '_')
parks = list(map(converter, parks_list))

data = [] #cria um array vazio para servir de base para o Data Frame

for park in parks:
    print(f"Buscando dados para o park {park}")
    BASE = 'https://en.wikipedia.org/w/index.php?title='+ urlparse.quote(park) +'&action=info'  
  
    park_url = urlopen(BASE)
    pk = BeautifulSoup(park_url, 'html.parser')
    tables = pk.find_all('table')
    infos_pk = {'nm_pk': park}
    for table in tables:
        if len(table) >1:
            rows = table.find_all('tr')
            for row in rows[0:]:
                cells = row.find_all('td')
                if (cells[0].text.strip())=='Date of page creation':
                    #print(parse(cells[1].text.strip()).date())
                    dic_1 = {
                        'dt_create_pg': cells[1].text.strip() #salva a data de criação da página
                    } #cria um dicionário que será adicionado ao array vazio
                    infos_pk.update(dic_1)
                if (cells[0].text.strip())=='Wikidata item ID':
                    dic_2 = {
                        'wikidata_id': cells[1].text.strip() #
                    } 
                    infos_pk.update(dic_2)
                if (cells[0].text.strip())=='Central description':
                    dic_3 = {
                        'dsc_central': cells[1].text.strip() 
                    } 
                    infos_pk.update(dic_3)
        data.append(infos_pk)
    
    sleep(3)


df = pd.DataFrame(data) #transforma o array "data" em um Data Frame
df
```
